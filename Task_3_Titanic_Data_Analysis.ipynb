**PART 2: TITANIC DATASET**

```python
file_path = "/content/drive/MyDrive/train.csv"

titanic = pd.read_csv(file_path)
```

**STEP 1: EXPLORATION**

**STEP 2: DATA CLEANING**

```python
# Fill missing Age with median
titanic["Age"].fillna(titanic["Age"].median(), inplace=True)

# Fill missing Embarked with mode
titanic["Embarked"].fillna(titanic["Embarked"].mode()[0], inplace=True)

# Drop Cabin column
titanic.drop(columns=["Cabin"], inplace=True, errors = "ignore")

# Remove duplicates
titanic.drop_duplicates(inplace=True)

# Check remaining missing values
titanic.isnull().sum()
```

**STEP 3: DATA ANALYSIS (GROUPBY)**

*1, Survival rate by gender*

```python
gender_survival = titanic.groupby("Sex")["Survived"].mean()
gender_survival
```

*2, Survival rate by class*

```python
class_survival = titanic.groupby("Pclass")["Survived"].mean()
class_survival
```

*3, Average age per class*

```python
avg_age = titanic.groupby("Pclass")["Age"].mean()
avg_age
```

*4. Survival rate by age group*

```python
titanic["AgeGroup"] = pd.cut(
    titanic["Age"],
    bins=[0, 12, 18, 35, 60, 100],
    labels=["Child", "Teen", "Adult", "Middle", "Senior"]
)

age_survival = titanic.groupby("AgeGroup")["Survived"].mean()
age_survival
```

**STEP 4: FILTERING**

*Female survivors*

```python
female_survivors = titanic[(titanic["Sex"] == "female") & (titanic["Survived"] == 1)]
female_survivors.head()
```

*Children survivors*

```python
children_survivors = titanic[(titanic["Age"] < 18) & (titanic["Survived"] == 1)]
children_survivors.head()
```

*1st class passengers survival rate*

```python
first_class_survival = titanic[titanic["Pclass"] == 1]["Survived"].mean()
print("First Class Survival Rate:", first_class_survival)
```

**STEP 5: INSIGHTS**

1. Who was more likely to survive? (Gender)

>> Females were more likely to survive than males.

The survival rate for female passengers was significantly higher.
This reflects the historical evacuation policy of “women and children first.”


2. Did class affect survival?

>> Yes, passenger class had a strong impact on survival.

1st class passengers had the highest survival rate.

3rd class passengers had the lowest survival rate.

This indicates that socio-economic status influenced survival chances, as higher-class passengers had better access to lifeboats and safer locations.

3. Were children prioritized?

>> Yes, children were prioritized during evacuation.

Children had a higher survival rate compared to many adult groups.
This suggests that vulnerable passengers were given priority.

4. What combination had the highest survival rate?

>> Female passengers in 1st class had the highest survival probability.

This shows that survival was influenced by a combination of factors:

Gender (female)
Class (1st class)
Age (younger passengers)

**Final Insight**

Survival on the Titanic was not random — it was strongly influenced by social hierarchy, gender-based priority, and age, with female first-class passengers having the highest chance of survival.
